```python
# 1. a.
# Define & calculate high level metrics like (Total Revenue, Total quantity, Total 
# products, Total categories, Total sellers, Total locations, Total channels, Total 
# payment methods etcâ€¦)

print('Total Revenue = ',round(oitems.price.sum(),3))
print('Total Quantity = ',oitems.shape[0])
print('Total Products = ',prod.product_id.count())
print('Total Categories = ',prod.product_category_name.nunique())
print('Total Locations = ',location.geolocation_zip_code_prefix.nunique())
print('Total Channels = ',seller.seller_id.nunique())
print('Total Payment methods = ',opayment.payment_type.nunique())
```

```python
#b how many new customers acquired every month
order['order_purchase_timestamp'] = pd.to_datetime(order.order_purchase_timestamp, format = '%m/%d/%Y %H:%M')
order['Year_month'] = order.order_purchase_timestamp.apply(lambda x: pd.Timestamp.strftime(x, '%Y-%m'))
```

```python
# merging customer and orders data

cust_order = pd.merge(left=cust,right=order,on='customer_id',how='left')
cust_order
```

```python
cust_month = cust_order.loc[:,['customer_unique_id','Year_month']]
```

```python
cust_month.drop_duplicates(subset='customer_unique_id',inplace=True)   ## dropping the duplicate customer unique id
```

```python
new_cust_monthly = cust_month.groupby('Year_month')[['customer_unique_id']].count()
new_cust_monthly
```

```python
new_cust_monthly.plot(kind='bar',figsize=(10,5))
plt.ylabel('No. of customers')
plt.title('New customers acquired every month')
plt.show()
```

```python
# 1. c. Understand the retention of customers on month on month basis

cust_monthly_count = cust_order.groupby('Year_month')[['customer_unique_id']].count().reset_index()
cust_monthly_count
```

```python
cust_monthly_count.customer_unique_id.sum()
```

```python
retented_cust = pd.merge(new_cust_monthly,cust_monthly_count,on='Year_month', how='left')
```

```python
retented_cust['count_retented_cust'] = retented_cust.customer_unique_id_y - retented_cust.customer_unique_id_x
retented_cust.head(2)
```

```python
retented_cust.plot(kind='bar',x='Year_month',y='count_retented_cust',figsize=(12,6))
plt.ylabel('count_of_retented_customers')
plt.title('retention of customers on month on month basis')
plt.show()
```

```python
# 1. d. How the revenues from existing/new customers on month on month basis.
order_item_month = pd.merge(oitems, order,how='left',on='order_id')
```

```python
order_item_month['net_price'] = order_item_month.price + order_item_month.freight_value
order_item_month
```

```python
revenue = order_item_month.groupby(['Year_month'])[['net_price']].sum()
revenue
```

```python
revenue.plot(kind='bar', y='net_price',ylabel = 'net_price',color='green',figsize=(12,6))
plt.title(' revenues from customers on month on month basis')
plt.show()
```

## Q.1 .e

```python
order_item_prod = pd.merge(order_item_month, prod,on='product_id',how='left')
```

```python
# 1. e. Understand the trends/seasonality of sales, quantity by category, location, month, week, day, time, channel, payment method e

order_item_prod['Month'] = order_item_prod.order_purchase_timestamp.apply(lambda x: pd.Timestamp.strftime(x,format='%b'))
order_item_prod
```

```python
# Month wise seasonality of products on sales and qty.
month_qty_sale = order_item_prod.groupby('Month').agg({'product_id':'count',
                                          'price':'sum'}).reset_index().rename(columns={'product_id':'Quantity',
                                                                                      'price':'sales'})
month_qty_sale
```

```python
# Seasonality of product category by sales and quantity
product_cat_qty_sale = order_item_prod.groupby('product_category_name').agg({'product_id':'count',
                                          'price':'sum'}).reset_index().rename(columns={'product_id':'Quantity',
                                                                                      'price':'sales'})
product_cat_qty_sale
```

```python
plt.plot(month_qty_sale['Month'],month_qty_sale['sales'],label='Qty')
```

```python
# 2. Customer /Seller segmentation
# a. Divide the customers into groups based on the revenue generated 

cust_order_oitem = pd.merge(cust_order, oitems, how='inner',on='order_id')
cust_order_oitem
```

```python
cust_order_oitem['net_price'] = cust_order_oitem.price + cust_order_oitem.freight_value
```

```python
cust_revenue = cust_order_oitem.groupby(by=['customer_id'])['net_price'].sum().reset_index()
cust_revenue.rename(columns={'net_price':'revenue_generated'},inplace=True)
cust_revenue
```

```python
# dividing / making groups of customers based on revenue
cust_revenue['groups'] = pd.cut(cust_revenue.revenue_generated,
       bins=np.arange(0, 15000, 1000),
       labels=['0-1000','1000-2000','2000-3000','3000-4000','4000-5000','5000-6000','6000-7000','7000-8000'
              ,'8000-9000','9000-10000','10000-11000','11000-12000','12000-13000','13000-14000']
        )
cust_revenue
```

```python
print(cust_revenue.revenue_generated.max(), cust_revenue.revenue_generated.min())
```

```python
# 5. Customer satisfaction towards category & product
# a. Which categories (top 10) are maximum rated & minimum rated?
order_item_prod_review = pd.merge(order_item_prod,oreview_rating,on='order_id',how='left')
order_item_prod_review
```

```python
cat_review = order_item_prod_review.groupby(by=['product_category_name'])[['review_score']].sum().reset_index()
max_cat_score = cat_review.sort_values(by=['review_score'],ascending=False).head(10).reset_index().drop(columns='index')

print('Top 10 maximum category rated:')
max_cat_score
```

```python
min_cat_score = cat_review.sort_values(by=['review_score'],ascending=True).head(10).reset_index().drop(columns='index')

print('Top 10 minimum category rated:')
min_cat_score
```

## Q. 5. b.

```python
# 5. b. Which products (top10) are maximum rated & minimum rated?

prod_review = order_item_prod_review.groupby(by=['product_id'])[['review_score']].sum().reset_index()
max_prod_score = prod_review.sort_values(by=['review_score'],ascending=False).head(10).reset_index().drop(columns='index')

print('Top 10 maximum category rated:')
max_prod_score
```

```python
min_prod_score = prod_review.sort_values(by=['review_score'],ascending=True).head(10).reset_index().drop(columns='index')

print('Top 10 minimum category rated:')
min_prod_score
```

```python
## Checking percentage of delivery cost and plotting distribution plot.

df_percentage_of_deliver_costs =\
                                oitems[['order_id','price','freight_value']].groupby('order_id').sum().reset_index()

df_percentage_of_deliver_costs['whole_cost']=\
                                df_percentage_of_deliver_costs[['price','freight_value']].sum(axis=1)

df_percentage_of_deliver_costs['%_of_cost']=round(df_percentage_of_deliver_costs['freight_value']
                                /df_percentage_of_deliver_costs['whole_cost'],3)*100

df_percentage_of_deliver_costs
```

```python
sns.distplot(df_percentage_of_deliver_costs['%_of_cost'])
plt.xlabel("% of delivery cost in order")
plt.show()
```
